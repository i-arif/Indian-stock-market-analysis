```python
stocks_names = ['reliance', 'tcs', 'infosys', 'hdfc_bank', 'wipro']
clean_data = {}

for name in stocks_names:
    df_temp = pd.read_csv(f'../data/{name}_2024.csv', skiprows=[1, 2], index_col=0, parse_dates=True)
    df_temp.index.name = 'Date'
    clean_data[name] = df_temp
    print(f"{name} cleaned — Shape: {df_temp.shape} ✅")
```

```python
for name, df_temp in clean_data.items():
    missing = df_temp.isnull().sum().sum()
    print(f"{name}: {missing} missing values")
```

```python
import os
os.makedirs('../data/cleaned', exist_ok=True)

for name, df_temp in clean_data.items():
    df_temp.to_csv(f'../data/cleaned/{name}_clean.csv')

print("All cleaned data saved! ✅")
```
