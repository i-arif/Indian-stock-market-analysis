```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
import warnings
warnings.filterwarnings('ignore')

print("All libraries imported successfully! ✅")
```

```python
stocks = {
    'Reliance': 'RELIANCE.NS',
    'TCS': 'TCS.NS',
    'Infosys': 'INFY.NS',
    'HDFC Bank': 'HDFCBANK.NS',
    'Wipro': 'WIPRO.NS'
}

print("Stocks defined! ✅")
```

```python
all_data = {}

for name, ticker in stocks.items():
    df = yf.download(ticker, start="2024-01-01", end="2024-12-31")
    all_data[name] = df
    print(f"{name} — {len(df)} rows downloaded ✅")
```

```python
import os
os.makedirs('../data', exist_ok=True)
print("Data folder ready! ✅")
```

```python
for name, df in all_data.items():
    df.to_csv(f"../data/{name.lower().replace(' ', '_')}_2024.csv")
    
print("All data saved to /data folder! ✅")
```
